NumPy를 사용하여 다음과 같은 배열을 생성해 보세요.

1. 크기가 4x3인 모든 값이 0인 배열
2. 크기가 3x3인 모든 값이 1인 배열
3. 0에서 20까지 5씩 증가하는 값으로 구성된 배열
4. 0과 1 사이의 4개의 등간격 값을 가진 배열
5. *크기가 2x2인 정규 분포로 구성된 배열*

```python
import numpy as np
zeros_array = np.zeros((4,3), dtype=int)
ones_array = np.ones((3,3), dtype=int)
range_array = np.arange(0, 20, 5)
linspace_array = np.linspace(0, 1, 4)
normal_array = np.random.normal(0, 1, (2,2))

print("크기가 4x3인 모든 값이 0인 배열: \n", zeros_array)
print("크기가 3x3인 모든 값이 1인 배열: \n", ones_array)
print("0에서 20까지 5씩 증가하는 값으로 구성된 배열: \n", range_array)
print("0과 1 사이의 4개의 등간격 값을 가진 배열: \n", linspace_array)
print("크기가 2x2인 정규 분포로 구성된 배열: \n", normal_array)
```

아래 배열에서 평균, 중앙값, 최빈값, 분산, 표준편차, 범위를 계산하세요.

```python
import numpy as np

arr = np.array([10, 20, 30, 30, 40, 50, 60])

# 최빈값 (scipy 없이)
values, counts = np.unique(arr, return_counts=True)
mode = values[np.argmax(counts)]

print(f"평균: {arr.mean():.4f}")
print(f"중앙값: {np.median(arr)}")
print(f"최빈값: {mode}")
print(f"분산: {arr.var():.4f}")
print(f"표준편차: {arr.std():.4f}")
print(f"범위: {arr.min()} ~ {arr.max()}")
```

아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.

```python
import numpy as np

scores_math =    [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
scores = np.array([scores_math, scores_english, scores_science])

# (3, 10) → (10, 3) : 한 행 = 한 학생의 3과목 점수
scores_T = scores.T

# 각 학생의 3과목 평균 (axis=1: 행 방향) → (10,)
student_avg = scores_T.mean(axis=1)

print("학생별 평균:", student_avg)
print("분산:", student_avg.var())
print("표준편차:", student_avg.std())
```

다음은 5개 지점에서 판매된 과일에 대한 월별 판매량 데이터입니다. 이 데이터는 각 지점의 판매량을 월별로 나타내며, 각 지점에서 판매된 사과, 바나나, 오렌지의 판매량이 기록되어 있습니다. 이 데이터를 사용하여 여러 가지 통계적 분석과 배열 조작을 수행하세요.

1. 각 지점에서 6개월 동안 판매된 **사과**, **바나나**, **오렌지**의 월별 평균 판매량을 계산하세요.
2. 각 월마다 전체 지점에서 가장 많이 판매된 과일을 찾아 판매량과 과일명을 출력하세요.
3. 6개월 동안 전체 지점에서 가장 많이 판매된 과일의 총량을 과일별로 구하세요. 어떤 과일이 가장 많이 판매되었는지 확인하세요.
4. 각 지점에서 판매량이 150개 이상인 데이터를 추출하고 그에 해당하는 지점과 과일을 출력하세요.

```python
import numpy as np

# 5개 지점의 월별 사과, 바나나, 오렌지 판매량 (5x3x6 배열)
# 5개의 지점, 3개의 과일(사과, 바나나, 오렌지), 6개월치 데이터
sales_data = np.array([
    # 지점 1
    [[120, 150, 200, 230, 180, 210],  # 사과 판매량
     [100, 140, 170, 190, 160, 180],  # 바나나 판매량
     [90, 120, 160, 140, 130, 150]],  # 오렌지 판매량

    # 지점 2
    [[80, 90, 110, 140, 130, 120],
     [60, 70, 90, 100, 110, 95],
     [50, 80, 100, 110, 90, 85]],

    # 지점 3
    [[130, 140, 160, 180, 170, 190],
     [110, 120, 130, 150, 140, 160],
     [100, 110, 120, 130, 120, 140]],

    # 지점 4
    [[100, 130, 150, 170, 160, 180],
     [90, 100, 110, 120, 130, 125],
     [80, 110, 140, 130, 120, 110]],

    # 지점 5
    [[90, 110, 130, 140, 120, 130],
     [70, 85, 100, 110, 105, 95],
     [60, 90, 110, 100, 90, 95]]
])


import numpy as np

fruit_names = np.array(['사과', '바나나', '오렌지'])
months = ['1월', '2월', '3월', '4월', '5월', '6월']

# 1. 각 지점에서 과일별 월별 평균 판매량 (axis=2: 월 방향 평균) → (5, 3)
branch_fruit_avg = sales_data.mean(axis=2)
print("=== 1. 지점별 과일 월평균 판매량 ===")
for i, branch in enumerate(branch_fruit_avg):
    print(f"지점 {i+1}: " + ", ".join(f"{fruit_names[j]} {branch[j]:.1f}" for j in range(3)))

# 2. 각 월마다 전체 지점 합산 후 가장 많이 팔린 과일 (axis=0: 지점 합) → (3, 6)
monthly_fruit_sum = sales_data.sum(axis=0)
print("\n=== 2. 월별 최다 판매 과일 ===")
for m in range(6):
    fruit_idx = np.argmax(monthly_fruit_sum[:, m])
    print(f"{months[m]}: {fruit_names[fruit_idx]} ({monthly_fruit_sum[fruit_idx, m]}개)")

# 3. 과일별 6개월 전체 지점 총 판매량 (지점·월 합) → (3,)
fruit_total = sales_data.sum(axis=(0, 2))
print("\n=== 3. 과일별 총 판매량 ===")
for j in range(3):
    print(f"{fruit_names[j]}: {fruit_total[j]}개")
print(f"→ 최다 판매 과일: {fruit_names[np.argmax(fruit_total)]} ({fruit_total.max()}개)")

# 4. 판매량 150개 이상 데이터 추출
print("\n=== 4. 판매량 150개 이상 데이터 ===")
branch_idx, fruit_idx, month_idx = np.where(sales_data >= 150)
for s, f, m in zip(branch_idx, fruit_idx, month_idx):
    print(f"지점 {s+1} | {fruit_names[f]} | {months[m]} | {sales_data[s, f, m]}개")
```
